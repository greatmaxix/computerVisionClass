```python
# data preparation

# unzipping the data
onePieceRoot = 'Manga109_released_2021_12_30/OnePiece/'

import os

allImagesPath = []
for dirs in os.listdir(onePieceRoot):
    # remove first item and last 3 items in a folder    
    for files in os.listdir(onePieceRoot + dirs)[:-3]:
        if files.endswith('.jpeg') and files != '0.jpeg':
            allImagesPath.append(onePieceRoot + dirs + '/' + files)

# copy all images to a new folder
import shutil

!rm -rf OnePieceData
!mkdir OnePieceData

for i in range(len(allImagesPath)):
    shutil.copy(allImagesPath[i], 'OnePieceData/' + str(i) + '.jpeg')

# create black and white copy of all images

!rm -rf OnePieceDataBW
!mkdir OnePieceDataBW

import cv2

for i in range(len(allImagesPath)):
    img = cv2.imread(allImagesPath[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('OnePieceDataBW/' + str(i) + '.jpeg', img)
```

```python
import os
import shutil

bwImagesPath = 'OnePieceDataBW'
coloredImagesPath = 'OnePieceData'

trainA = []
trainB = []
valA = []
valB = []

!rm -rf onePieceSplitted
!mkdir -p onePieceSplitted
!mkdir -p onePieceSplitted/trainA
!mkdir -p onePieceSplitted/trainB
!mkdir -p onePieceSplitted/valA
!mkdir -p onePieceSplitted/valB

bwImagesFiles = os.listdir(bwImagesPath)
coloredImagesFiles = os.listdir(coloredImagesPath)

# copy half of the images to trainA and the other half to trainB
for i in range(len(coloredImagesFiles)):
    # 80% of the data for training
    if i < len(coloredImagesFiles) * 0.8:
        shutil.copy(bwImagesPath + '/' + bwImagesFiles[i], 'onePieceSplitted/trainA/' + bwImagesFiles[i])
        shutil.copy(coloredImagesPath + '/' + coloredImagesFiles[i], 'onePieceSplitted/trainB/' + coloredImagesFiles[i])
    else:
        shutil.copy(bwImagesPath + '/' + bwImagesFiles[i], 'onePieceSplitted/valA/' + bwImagesFiles[i])
        shutil.copy(coloredImagesPath + '/' + coloredImagesFiles[i], 'onePieceSplitted/valB/' + coloredImagesFiles[i])
```
